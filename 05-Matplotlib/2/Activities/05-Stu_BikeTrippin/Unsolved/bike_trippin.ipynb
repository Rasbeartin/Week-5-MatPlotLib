* Split up your data into groups based upon the gender column
    * NOTE: There will be a garbage row with a gender of 'stoptime' which you will have to remove!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_to_load = '../Resources/trip.csv'
trip_df = pd.read_csv(file_to_load)
trip_df.head()
```

```python
gender_group_df = trip_df.groupby('gender')
trip_dur = gender_group_df['tripduration'].count()
#gender_trip_df = pd.DataFrame(trip_total)
#gender_trip_df
trip_dur_filter = trip_dur.drop(index = 'stoptime')
```

```python
trip_dur_filter.plot(kind = 'bar', title='Trip Duration to Gender')
```
